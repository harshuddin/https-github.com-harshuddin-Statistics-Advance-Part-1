Cumulative Distribution Functions (CDFs):

2. Cumulative: CDFs provide the cumulative probability up to a given point.

Key Differences:
1. Cumulative vs. non-cumulative: CDFs are cumulative, while PDFs are non-cumulative.
2. Interpretation: PDFs describe the probability density at a point, while CDFs describe the cumulative probability up to a point.

Relationship Between CDFs and PDFs:

Importance of CDFs:

Q8.  What is the binomial distribution, and how is it used in probability?

Answer>>The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials, where each trial has a constant probability of success.

Characteristics:
1. Fixed number of trials: The binomial distribution models a fixed number of trials, denoted by n.
2. Independent trials: Each trial is independent of the others.
3. Constant probability of success: The probability of success, denoted by p, is constant for each trial.
4. Two possible outcomes: Each trial has two possible outcomes, typically labeled as success (1) and failure (0).

Probability Mass Function:
The probability mass function of the binomial distribution is given by:

P(X = k) = (n choose k) * p^k * (1-p)^(n-k)

where k is the number of successes, n is the number of trials, and p is the probability of success.

Properties:
1. Mean: The mean of the binomial distribution is np.
2. Variance: The variance of the binomial distribution is np(1-p).

Q11. What are the characteristics of a normal distribution?
Answer>> A normal distribution, also known as a Gaussian distribution or bell curve, is a continuous probability distribution that is commonly observed in many natural and social phenomena. The characteristics of a normal distribution are:

Characteristics:
1. Symmetry: The normal distribution is symmetric about its mean, meaning that the left and right sides of the distribution are mirror images of each other.
2. Bell-shaped: The normal distribution has a bell-shaped curve, with the majority of the data points clustered around the mean and tapering off gradually towards the extremes.
3. Mean, median, and mode: In a normal distribution, the mean, median, and mode are all equal.
4. Continuous: The normal distribution is a continuous distribution, meaning that it can take on any value within a given range.

Parameters:
1. Mean (μ): The mean of a normal distribution determines the location of the distribution.
2. Standard deviation (σ): The standard deviation of a normal distribution determines the spread or dispersion of the distribution.

Properties:
1. 68-95-99.7 rule: About 68% of the data points fall within 1 standard deviation of the mean, about 95% fall within 2 standard deviations, and about 99.7% fall within 3 standard deviations.
2. Additivity: The sum of independent normal random variables is also normally distributed.

Real-World Applications:
1. Natural phenomena: Normal distributions are commonly observed in natural phenomena, such as the heights of people, the weights of objects, and the errors in measurement.
2. Social sciences: Normal distributions are used to model many social science variables, such as IQ scores, income, and test scores.
3. Finance: Normal distributions are used to model stock prices, returns on investment, and other financial variables.

Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Answer>> The Central Limit Theorem (CLT) is a fundamental concept in statistics that describes the behavior of the sampling distribution of the mean.

Definition:
The CLT states that, given certain conditions, the sampling distribution of the mean will be approximately normally distributed, even if the underlying population distribution is not normal.

Conditions:
1. Random sampling: The sample is selected randomly from the population.
2. Independence: The observations in the sample are independent of each other.
3. Sufficient sample size: The sample size is sufficiently large (usually n ≥ 30).

Importance:
1. Approximation to normal distribution: The CLT allows us to approximate the sampling distribution of the mean to a normal distribution, which facilitates statistical inference.
2. Hypothesis testing: The CLT is used in hypothesis testing to determine the probability of observing a sample mean or proportion.
3. Confidence intervals: The CLT is used to construct confidence intervals for population means and proportions.

Implications:
1. Robustness: The CLT provides a robust framework for statistical inference, even when the underlying population distribution is not normal.
2. Generalizability: The CLT allows us to generalize findings from a sample to the larger population.

Applications:
1. Statistical inference: The CLT is used in statistical inference, including hypothesis testing and confidence intervals.
2. Data analysis: The CLT is used in data analysis to identify patterns and trends in data.
3. Quality control: The CLT is used in quality control to monitor and control processes.

Q15. What is the application of Z statistics in hypothesis testing?

Answer>> Z-statistics are used in hypothesis testing to determine the probability of observing a sample mean or proportion, given a known population mean and standard deviation.

Application:
1. Testing population means: Z-statistics are used to test hypotheses about population means, such as comparing a sample mean to a known population mean.
2. Testing population proportions: Z-statistics can be used to test hypotheses about population proportions, such as comparing a sample proportion to a known population proportion.

Calculation:
The Z-statistic is calculated using the following formula:

Z = (x̄ - μ) / (σ / √n)

where x̄ is the sample mean, μ is the population mean, σ is the population standard deviation, and n is the sample size.

Interpretation:
1. P-value: The Z-statistic is used to calculate the p-value, which represents the probability of observing a sample mean or proportion at least as extreme as the one observed, assuming the null hypothesis is true.
2. Rejecting the null hypothesis: If the p-value is below a certain significance level (usually 0.05), the null hypothesis is rejected, indicating that the sample mean or proportion is statistically significantly different from the population mean or proportion.

Importance:
1. Statistical inference: Z-statistics are essential for statistical inference, allowing researchers to make informed decisions about population parameters based on sample data.
2. Hypothesis testing: Z-statistics are widely used in hypothesis testing, enabling researchers to test hypotheses about population means and proportions.

Answer>>A Z-score, also known as a standard score, is a measure of how many standard deviations an observation is away from the mean of a normal distribution.

Calculation:
The Z-score is calculated using the following formula:

Z = (X - μ) / σ

Interpretation:

```python
#Q3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    - x (int): The value (0 or 1).
    - p (float): The probability of success (0 <= p <= 1).

    Returns:
    - float: The probability P(X = x).
    """
    if not 0 <= p <= 1:
        raise ValueError("Probability p must be between 0 and 1.")

    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0.0

```

```python
#Q6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discreteuniform distribution.
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(a, b):
    """Compute and plot the CDF of a discrete uniform distribution from a to b."""
    if a > b:
        raise ValueError("Lower bound 'a' must be less than or equal to upper bound 'b'.")

    # Possible outcomes
    x = np.arange(a, b + 1)
    # Equal probabilities
    prob = 1 / (b - a + 1)
    # CDF values
    cdf = np.cumsum([prob] * len(x))

    # Plotting
    plt.figure(figsize=(8, 5))
    plt.step(x, cdf, where='post', label="CDF", linewidth=2)
    plt.xticks(x)
    plt.xlabel('x')
    plt.ylabel('F(x)')
    plt.title(f'CDF of Discrete Uniform Distribution (a={a}, b={b})')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
a = 1
b = 6
discrete_uniform_cdf(a, b)
```

```python
#Q7.> Generate a continuous uniform distribution using NumPy and visualize it.
import numpy as np
import matplotlib.pyplot as plt

def plot_continuous_uniform(a, b, num_samples=1000):
    if a >= b:
        raise ValueError("Lower bound 'a' must be less than upper bound 'b'.")

    # Generate samples from uniform distribution
    data = np.random.uniform(low=a, high=b, size=num_samples)

    # Sort data for CDF
    sorted_data = np.sort(data)
    cdf = np.arange(1, num_samples + 1) / num_samples

    # Plotting
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    # PDF - Histogram
    axs[0].hist(data, bins=30, density=True, color='skyblue', edgecolor='black')
    axs[0].set_title(f'PDF of Uniform({a}, {b})')
    axs[0].set_xlabel('x')
    axs[0].set_ylabel('Density')
    axs[0].grid(True)

    # CDF
    axs[1].plot(sorted_data, cdf, marker='.', linestyle='none', color='darkorange')
    axs[1].set_title(f'Empirical CDF of Uniform({a}, {b})')
    axs[1].set_xlabel('x')
    axs[1].set_ylabel('F(x)')
    axs[1].grid(True)

    plt.tight_layout()
    plt.show()

# Example usage
plot_continuous_uniform(a=0, b=10, num_samples=1000)
```

```python
#Q9.Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Convert to NumPy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot histogram of Z-scores
    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=30, color='cornflowerblue', edgecolor='black', alpha=0.7, density=True)
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid(True)
    plt.axvline(x=0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores

# Example usage:
# Generate sample data from normal distribution
sample_data = np.random.normal(loc=50, scale=10, size=1000)
z_scores = calculate_and_plot_z_scores(sample_data)
```

```python
#Q10.Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

def demonstrate_clt(sample_size=30, num_samples=1000, distribution='exponential'):
    # Step 1: Define original non-normal distribution
    if distribution == 'exponential':
        population_data = np.random.exponential(scale=2.0, size=100000)
    elif distribution == 'uniform':
        population_data = np.random.uniform(low=0, high=10, size=100000)
    else:
        raise ValueError("Supported distributions: 'exponential', 'uniform'")

    # Step 2: Collect sample means
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population_data, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))

    # Step 3: Plot the results
    plt.figure(figsize=(12, 5))

    # Original population distribution
    plt.subplot(1, 2, 1)
    plt.hist(population_data, bins=50, color='salmon', edgecolor='black', alpha=0.7, density=True)
    plt.title(f'Original {distribution.capitalize()} Distribution')
    plt.xlabel('Value')
    plt.ylabel('Density')

    # Distribution of sample means
    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=30, color='seagreen', edgecolor='black', alpha=0.7, density=True)
    plt.title(f'Distribution of Sample Means\n(n={sample_size}, samples={num_samples})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')

    plt.tight_layout()
    plt.show()

# Example usage
demonstrate_clt(sample_size=50, num_samples=1000, distribution='exponential')
```

```python
#Q13.Generate random variables and calculate their corresponding probabilities using the binomial distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def simulate_binomial(n=10, p=0.5, num_samples=1000, plot=True):
    # Step 1: Generate random variables
    data = np.random.binomial(n=n, p=p, size=num_samples)

    # Step 2: Calculate PMF (probability of each possible outcome)
    x = np.arange(0, n + 1)
    probabilities = binom.pmf(x, n, p)

    # Print sample of outcomes and probabilities
    print("Sample of outcomes:", data[:10])
    print("PMF values for all possible outcomes (k = 0 to n):")
    for k, prob in zip(x, probabilities):
        print(f"P(X = {k}) = {prob:.4f}")

    # Step 3: Optional plot
    if plot:
        plt.figure(figsize=(8, 5))
        plt.bar(x, probabilities, color='skyblue', edgecolor='black', label='PMF')
        plt.title(f'Binomial Distribution PMF (n={n}, p={p})')
        plt.xlabel('Number of Successes (k)')
        plt.ylabel('Probability')
        plt.xticks(x)
        plt.grid(axis='y')
        plt.legend()
        plt.tight_layout()
        plt.show()

    return data, probabilities

# Example usage
simulate_binomial(n=10, p=0.3, num_samples=1000)
```

```python
#Q14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score_and_plot(data_point, data):
    # Step 1: Calculate mean and std of the data
    mean = np.mean(data)
    std_dev = np.std(data)

    # Step 2: Calculate Z-score
    z_score = (data_point - mean) / std_dev
    print(f"Data point: {data_point}")
    print(f"Mean of data: {mean:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"Z-score: {z_score:.2f}")

    # Step 3: Plot standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, loc=0, scale=1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, color='navy', label='Standard Normal PDF')
    plt.fill_between(x, y, where=(x < z_score), color='skyblue', alpha=0.4, label='Area Left of Z')
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
    plt.title('Z-score on Standard Normal Distribution')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return z_score

# Example usage:
# Simulated dataset
data = np.random.normal(loc=100, scale=15, size=1000)
# Data point to evaluate
data_point = 130

# Run
z_score = calculate_z_score_and_plot(data_point, data)
```

```python
#Q15. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    # Step 1: Sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Step 2: Compute Z statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Step 3: Compute p-value (two-tailed)
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Step 4: Result
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Step 5: Conclusion
    if p_value < alpha:
        print("Result: Reject the null hypothesis (significant difference)")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference)")

# --- Example usage ---
# Simulate a sample from a normal distribution
np.random.seed(42)
sample = np.random.normal(loc=105, scale=15, size=50)  # Sample from a population with mean ≈ 105

# Assume known population mean and std deviation
known_population_mean = 100
known_population_std = 15

z_test(sample, known_population_mean, known_population_std, alpha=0.05)
```

```python
#Q16.Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence_level=0.95):
    # Step 1: Sample statistics
    sample_mean = np.mean(data)
    sample_size = len(data)
    sample_std_dev = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for sample std)

    # Step 2: t-critical value for the given confidence level
    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)  # Two-tailed t-distribution

    # Step 3: Margin of error
    margin_of_error = t_critical * (sample_std_dev / np.sqrt(sample_size))

    # Step 4: Confidence interval
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

    # Step 5: Print result
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Standard Deviation: {sample_std_dev:.2f}")
    print(f"Confidence Interval: {confidence_interval}")
    print(f"t-critical value: {t_critical:.4f}")

    return confidence_interval

# Example usage:
# Simulated data (e.g., scores of a class)
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=50)  # Mean = 100, Std = 15, n = 50

confidence_interval = calculate_confidence_interval(data, confidence_level=0.95)
```

```python
#Q17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence_level=0.95):
    # Step 1: Sample statistics
    sample_mean = np.mean(data)
    sample_size = len(data)
    sample_std_dev = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for sample std)

    # Step 2: t-critical value for the given confidence level
    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)  # Two-tailed t-distribution

    # Step 3: Margin of error
    margin_of_error = t_critical * (sample_std_dev / np.sqrt(sample_size))

    # Step 4: Confidence interval
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

    # Step 5: Print result
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Standard Deviation: {sample_std_dev:.2f}")
    print(f"Confidence Interval: {confidence_interval}")
    print(f"t-critical value: {t_critical:.4f}")

    return confidence_interval

# --- Simulate data from a normal distribution ---
np.random.seed(42)  # For reproducibility
population_mean = 100  # Known population mean
population_std = 15    # Known population standard deviation
sample_size = 50       # Sample size

# Generate data from the normal distribution
data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Calculate and interpret the confidence interval for the sample data
confidence_interval = calculate_confidence_interval(data, confidence_level=0.95)
```

```python
#Q18.Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mu=0, sigma=1, x_range=(-5, 5), num_points=1000):
    # Step 1: Generate x values from the specified range
    x = np.linspace(x_range[0], x_range[1], num_points)

    # Step 2: Calculate the PDF for these x values
    y = norm.pdf(x, loc=mu, scale=sigma)

    # Step 3: Plot the normal distribution PDF
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, color='blue', label=f'Normal Distribution\n(μ={mu}, σ={sigma})', lw=2)
    plt.fill_between(x, y, color='skyblue', alpha=0.4)

    # Adding labels and title
    plt.title('Probability Density Function (PDF) of Normal Distribution')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.axvline(mu, color='red', linestyle='--', label=f'Mean (μ={mu})')
    plt.legend(loc='best')
    plt.grid(True)
    plt.tight_layout()

    # Step 4: Show the plot
    plt.show()

# Example usage: Plot the PDF for a normal distribution with μ=0 and σ=1
plot_normal_pdf(mu=0, sigma=1, x_range=(-5, 5))
```

```python
#Q19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lmbda=3, max_k=10):
    # Step 1: Generate k values (possible number of events)
    k = np.arange(0, max_k + 1)

    # Step 2: Calculate the CDF for each k value
    cdf_values = poisson.cdf(k, mu=lmbda)

    # Step 3: Plot the Poisson CDF
    plt.figure(figsize=(8, 5))
    plt.step(k, cdf_values, where='post', color='blue', label=f'Poisson CDF (λ={lmbda})')
    plt.fill_between(k, cdf_values, color='skyblue', alpha=0.4)

    # Adding labels and title
    plt.title(f'Cumulative Distribution Function (CDF) of Poisson Distribution\n(λ={lmbda})')
    plt.xlabel('Number of Events (k)')
    plt.ylabel('Cumulative Probability')
    plt.grid(True)
    plt.xticks(k)
    plt.legend(loc='best')

    # Step 4: Show the plot
    plt.tight_layout()
    plt.show()

# Example usage: Plot the CDF for a Poisson distribution with λ = 3
plot_poisson_cdf(lmbda=3, max_k=10)
```

```python
#Q22.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
import numpy as np

def calculate_range_and_iqr(data):
    # Step 1: Calculate the range (max - min)
    data_range = np.max(data) - np.min(data)

    # Step 2: Calculate the IQR (Q3 - Q1)
    Q1 = np.percentile(data, 25)  # 25th percentile
    Q3 = np.percentile(data, 75)  # 75th percentile
    iqr = Q3 - Q1

    # Print results
    print(f"Range: {data_range:.4f}")
    print(f"Interquartile Range (IQR): {iqr:.4f}")

    return data_range, iqr

# --- Example Usage ---
# Simulate a dataset from a normal distribution (mean = 100, std = 15, n = 1000)
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=1000)

# Calculate and display range and IQR
range_value, iqr_value = calculate_range_and_iqr(data)
```

```python
#Q23. Implement Z-score normalization on a dataset and visualize its transformation.
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    # Step 1: Calculate the mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Step 2: Apply Z-score normalization
    z_scores = (data - mean) / std_dev

    return z_scores, mean, std_dev

def visualize_transformation(original_data, normalized_data):
    # Step 3: Visualize the original and normalized data
    plt.figure(figsize=(10, 6))

    # Plot the original data
    plt.subplot(1, 2, 1)
    plt.hist(original_data, bins=30, color='blue', alpha=0.7)
    plt.title('Original Data Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    # Plot the normalized data
    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=30, color='orange', alpha=0.7)
    plt.title('Normalized Data Distribution (Z-score)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

# --- Example Usage ---
# Simulate a dataset from a normal distribution (mean = 100, std = 15, n = 1000)
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=1000)

# Apply Z-score normalization
normalized_data, mean, std_dev = z_score_normalization(data)

# Print the mean and standard deviation of the original data
print(f"Original Mean: {mean:.2f}")
print(f"Original Standard Deviation: {std_dev:.2f}")

# Visualize the transformation
visualize_transformation(data, normalized_data)
```

```python
#Q24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_and_kurtosis(data):
    # Calculate skewness and kurtosis using scipy
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)

    # Print results
    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis: {data_kurtosis:.4f}")

    return data_skewness, data_kurtosis

# --- Example Usage ---
# Simulate a dataset from a normal distribution (mean = 100, std = 15, n = 1000)
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=1000)

# Calculate and display skewness and kurtosis
skewness_value, kurtosis_value = calculate_skewness_and_kurtosis(data)
```
